```python
import sys
main_path = "../.."
sys.path.append(main_path)

# importing modules
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from src.scripts.model_and_evaluation import ModelAndEvaluation
import pickle

import warnings
warnings.simplefilter('ignore')

# LungFlag model
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model_and_evaluation = ModelAndEvaluation()
```

```python
pipe_lungflag = Pipeline([['sc', StandardScaler()], ['clf', AdaBoostClassifier(
                 base_estimator = DecisionTreeClassifier(), n_estimators=1000, random_state=0)]])
```

```python
df_path = f"{main_path}/data/processed/dataframes"
targets = pd.read_csv(f"{df_path}/targets.csv").values.ravel()
binary_data = pd.read_csv(f"{df_path}/binary_data_is_live.csv").values.ravel()
data_live_month = pd.read_csv(f"{df_path}/data_live_month.csv").values.ravel()

print((binary_data))
```

```python
ohe_di = pd.read_csv(f"{df_path}/ohe_di.csv")
cf_metrics_score_ohe_is_live = model_and_evaluation.cross_val(pipe_lungflag, ohe_di.values, binary_data, grid_search=False)
print(cf_metrics_score_ohe_is_live)
```

### OHE -> DI

```python
ohe_di = pd.read_csv(f"{df_path}/ohe_di.csv")
cf_metrics_score_ohe_di = model_and_evaluation.cross_val(pipe_lungflag, ohe_di.values, targets, grid_search=False)
cf_metrics_score_ohe_di_is_live = model_and_evaluation.cross_val(pipe_lungflag, ohe_di.values, binary_data, grid_search=False)
cf_metrics_score_ohe_di_live_month = model_and_evaluation.cross_val(pipe_lungflag, ohe_di[0:70].values, data_live_month, grid_search=False)
print(cf_metrics_score_ohe_di)
```

```python
ohe_di = pd.read_csv(f"{df_path}/ohe_di.csv")
cf_metrics_score_ohe_di = model_and_evaluation.cross_val(pipe_lungflag, ohe_di.values, targets, grid_search=False)
cf_metrics_score_ohe_di_is_live = model_and_evaluation.cross_val(pipe_lungflag, ohe_di.values, binary_data, grid_search=False)
cf_metrics_score_ohe_di_live_month = model_and_evaluation.cross_val(pipe_lungflag, ohe_di[0:70].values, data_live_month, grid_search=False)
```

### OHE -> DI -> FI

```python
ohe_di_fi = pd.read_csv(f"{df_path}/ohe_di_fi.csv")
cf_metrics_score_ohe_di_fi = model_and_evaluation.cross_val(pipe_lungflag, ohe_di_fi.values, targets, grid_search=False)
cf_metrics_score_ohe_di_fi_is_live = model_and_evaluation.cross_val(pipe_lungflag, ohe_di_fi.values, binary_data, grid_search=False)
cf_metrics_score_ohe_di_fi_live_month = model_and_evaluation.cross_val(pipe_lungflag, ohe_di_fi[0:70].values, data_live_month, grid_search=False)
```

```python
from sklearn.model_selection import cross_val_score
result = cross_val_score(estimator=pipe_lungflag, X=ohe_di_fi.values, y=targets, cv=10, scoring='f1_micro')
result_is_live = cross_val_score(estimator=pipe_lungflag, X=ohe_di_fi.values, y=binary_data, cv=10, scoring='f1_micro')
result_live_month = cross_val_score(estimator=pipe_lungflag, X=ohe_di_fi[0:70].values, y=data_live_month, cv=10, scoring='f1_micro')

```

```python
result.mean()
result_is_live.mean()
result_live_month.mean()
```
